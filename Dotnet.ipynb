# Get File List with listdir

```python
from os import listdir
from os.path import isfile, join
```

```python
samples_folder = "samples/"
samples = [f for f in listdir(samples_folder) if isfile(join(samples_folder, f))]
```

```python
print(samples)
```

# Get File List with glob

```python
import glob
```

```python
files = glob.glob("samples/*")
```

```python
print(files)
```

# Install and Import dotnetfile

```python
from dotnetfile import DotNetPE
```

**make sure to instantiate files as dotnet file objects first with DotNetPE()**

```python
for file in files:
    sample_file = DotNetPE(file)
    print(sample_file.get_runtime_target_version())
```

```python
for file in files:
    sample_file = DotNetPE(file)
    stream_names = sample_file.get_stream_names()
    print(file)
    for stream_name in stream_names:
        print(f"\t{stream_name}")
    #print(sample_file.get_number_of_streams())
```

## Get Stream Strings

```python
for file in files:
    sample_file = DotNetPE(file)
    print(f"#Strings stream strings for: {file}")
    strings_stream_strings = sample_file.get_strings_stream_strings()
    for string in strings_stream_strings:
        print(f'\t{string}')
    print()
    print(f"#US stream strings for: {file}")
    us_stream_strings = sample_file.get_user_stream_strings()
    for string in us_stream_strings:
        print(f'\t{string}')
```

## Look for Anti-Analysis Techniques

```python
for file in files:
    print(f"Anti analysis tricks for {file}")
    print(f"\t.NET data directory hidden in PE header: {sample_file.AntiMetadataAnalysis.is_dotnet_data_directory_hidden}")
    print(f"\tHas extra data at the end of the metadata header: {sample_file.AntiMetadataAnalysis.has_metadata_table_extra_data}")
    print(f"\tHas fake types that reference each other: {sample_file.AntiMetadataAnalysis.has_self_referenced_typeref_entries}")
    print(f"\tHas invalid entries in TypeRef table: {sample_file.AntiMetadataAnalysis.has_invalid_typeref_entries}")
    print(f"\tHas fake data streams: {sample_file.AntiMetadataAnalysis.has_fake_data_streams}")
    print(f"\tHas more than one row in Module table: {sample_file.AntiMetadataAnalysis.module_table_has_multiple_rows}")
    print(f"\tHas more than one row in Assembly table: {sample_file.AntiMetadataAnalysis.assembly_table_has_multiple_rows}")
    print(f"\tHas invalid entries in #Strings stream: {sample_file.AntiMetadataAnalysis.has_invalid_strings_stream_entries}\n")
```

## Print Metadata Tables

```python
for file in files:
    sample_file = DotNetPE(file)
    print(f"Existent metadata tables for {file}")
    available_tables = sample_file.existent_metadata_tables()
    for table in available_tables:
        print(f"\t{table}")
```

## Print References

```python
for file in files:
    sample_file = DotNetPE(file)
    print(f"All references for: {file}")
    references = sample_file.get_all_references()
    for ref in references:
        print(f"\t{ref}")
```
