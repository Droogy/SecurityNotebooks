## Extract Phishing Links
- Create 2 folders, "Attachments" and "Mail"
- Drop phishing mail in "Mail" folder and then run through this script

```python
import os
import re
from independentsoft.msg import Message
from independentsoft.msg import Attachment
```

Create list of mail files to iterate over

```python
msg_mail_dir = "Mail/"
msg_files = []

for file in os.listdir(msg_mail_dir):
    f = os.path.join(file)
    if os.path.isfile(f"Mail/{f}"):
        msg_files.append(f)
```

Create list of all e-mail body text

```python
email_bodies = []

for file in msg_files:
    message = Message(f"Mail/{file}")
    email_bodies.append(message.body)
    for i in range(len(message.attachments)):
        attachment = message.attachments[i]
        attachment.save(f"Attachments/{attachment.file_name}")
```

Extract URLs from combined e-mail bodies list

```python
phishing_links = []
for body in email_bodies:
    m = re.search("(<.+/v3/__)(http.+)(__;.+>)", body, flags=re.M)
    m2 = re.search("<http.+", body, flags=re.M)
    if m:
        phishing_links.append(m.group(2))
    elif m2:
        phishing_links.append(m2.group())
    else:
        continue
```
